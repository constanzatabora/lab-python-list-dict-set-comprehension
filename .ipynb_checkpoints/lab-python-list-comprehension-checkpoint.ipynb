````python
#Exercise 1:
def initialize_inventory(products):
    return {product: int(input(f"Enter the quantity for {product}: ")) for product in products}
````

````python
def get_customer_orders():
    customer_orders = set()
    while True:
        order = input("Please enter the name of a product to order (choose from t-shirt, mug, hat, book, keychain): ").lower()
        if order in ["t-shirt", "mug", "hat", "book", "keychain"]:
            customer_orders.add(order)
        else:
            print("Invalid product. Please choose from the available products.")
        
        another_product = input("Do you want to add another product? (yes/no): ").lower()
        if another_product != 'yes':
            break
    return customer_orders
````

````python
def update_inventory(customer_orders, inventory):
    return {product: (inventory[product] - 1 if product in customer_orders and inventory[product] > 0 else inventory[product]) 
            for product in inventory}
````

````python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered
````

````python
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
````

````python
# List of available products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize inventory
inventory = initialize_inventory(products)

# Get customer orders
customer_orders = get_customer_orders()

# Calculate order statistics
order_statistics = calculate_order_statistics(customer_orders, products)

# Print order statistics
print_order_statistics(order_statistics)

# Update inventory based on customer orders
inventory = update_inventory(customer_orders, inventory)

# Print the updated inventory
print_updated_inventory(inventory)
````

````python
# List of available products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Function to initialize the inventory
def initialize_inventory(products):
    return {product: int(input(f"Enter the quantity for {product}: ")) for product in products}

# Function to get customer orders using comprehension
def get_customer_orders(products):
    try:
        num_orders = int(input("How many products would you like to order? "))
    except ValueError:
        print("Please enter a valid number.")
        return set()

    customer_orders = {input(f"Enter the name of product {i + 1} (choose from t-shirt, mug, hat, book, keychain): ").lower()
                       for i in range(num_orders) 
                       if input(f"Enter the name of product {i + 1}: ").lower() in products}

    return customer_orders

# Function to update the inventory
def update_inventory(customer_orders, inventory):
    return {product: (inventory[product] - 1 if product in customer_orders and inventory[product] > 0 else inventory[product]) 
            for product in inventory}

# Function to calculate order statistics
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered

# Function to print order statistics
def print_order_statistics(order_statistics):
    total_products_ordered, percentage_ordered = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_ordered:.2f}%")

# Function to print updated inventory
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# Sequence of calling the functions

# Step 1: Initialize inventory
inventory = initialize_inventory(products)

# Step 2: Get customer orders
customer_orders = get_customer_orders(products)

# Step 3: Calculate order statistics
order_statistics = calculate_order_statistics(customer_orders, products)

# Step 4: Print order statistics
print_order_statistics(order_statistics)

# Step 5: Update inventory based on customer orders
inventory = update_inventory(customer_orders, inventory)

# Step 6: Print updated inventory
print_updated_inventory(inventory)
````

````python
#Exercise 4
def update_inventory(customer_orders, inventory):
    # Update the inventory by subtracting 1 from the quantity for each product ordered
    updated_inventory = {product: (inventory[product] - 1 if product in customer_orders and inventory[product] > 0 else inventory[product])
                         for product in inventory}
    
    # Remove products with a quantity of 0
    return {product: quantity for product, quantity in updated_inventory.items() if quantity > 0}
````

````python
#Exercise 5
# List of available products
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Function to initialize the inventory
def initialize_inventory(products):
    return {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}

# Function to get customer orders using comprehension
def get_customer_orders(products):
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {input("Enter the name of a product that a customer wants to order: ").lower()
                       for _ in range(num_orders) if input("Enter the name of a product that a customer wants to order: ").lower() in products}
    return customer_orders

# Updated function to update the inventory
def update_inventory(customer_orders, inventory):
    updated_inventory = {product: (inventory[product] - 1 if product in customer_orders and inventory[product] > 0 else inventory[product])
                         for product in inventory}
    return {product: quantity for product, quantity in updated_inventory.items() if quantity > 0}

# Function to calculate order statistics
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered

# Function to print order statistics
def print_order_statistics(order_statistics):
    total_products_ordered, percentage_ordered = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage_ordered:.1f}")

# Function to print updated inventory
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# Function to calculate total price
def calculate_total_price(customer_orders):
    total_price = sum(float(input(f"Enter the price of {product}: ")) for product in customer_orders)
    return total_price

# Main program execution
# Step 1: Initialize inventory
inventory = initialize_inventory(products)

# Step 2: Get customer orders
customer_orders = get_customer_orders(products)

# Step 3: Calculate order statistics
order_statistics = calculate_order_statistics(customer_orders, products)

# Step 4: Print order statistics
print_order_statistics(order_statistics)

# Step 5: Update inventory based on customer orders
inventory = update_inventory(customer_orders, inventory)

# Step 6: Print updated inventory
print_updated_inventory(inventory)

# Step 7: Calculate and print total price
total_price = calculate_total_price(customer_orders)
print(f"\nTotal Price: {total_price:.1f}")
````
